```python
import pandas as pd 
import numpy as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv('spain_menorca.csv')
```

```python
# a) Los registros de 5 "host name" diferentes
filtro1 = df[df["host_name"].isin(df["host_name"].unique()[:5])]
filtro1
```

```python
#filtro b 
df['host_since'] = pd.to_datetime(df['host_since'])

# Filtrar los registros con 'host_since' menor a 2009
filtro2 = df[df['host_since'] < '2019-01-01']
filtro2
```

```python
#filtro cresponden a mas tardar un dia
filtro3 = df[(df['host_response_time'] == 'within a day') | (df['host_response_time'] == 'within an hour') | (df['host_response_time'] == 'within a few hours')]
filtro3
```

```python
# d) Los registros que tengan "al menos 5 accommodates"
filtro4 = df[df["accommodates"] >= 5]
filtro4
```

```python
#FILTR0e
# Filtrar los registros donde 'room_type' es 'Entire home' o 'Private room'
filtro5 = df[(df['room_type'] == 'Entire home') | (df['room_type'] == 'Private room')]
filtro5
```

```python
#filtro f tengan a lo mucho con 4 camas

df['bedrooms'] = pd.to_numeric(df['bedrooms'])

# Filtrar los registros donde 'bedrooms' es menor a 5
filtro6 = df[df['bedrooms'] < 5]

filtro6
```

```python
# g) Los registros que "tengan 1 baño o que sean superhost" pero con la condición de que "todos sean instant_bookable"
filtro7 = df[
    ((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) 
    & (df["instant_bookable"] == "t")
]
filtro7
```

```python
# h) Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_rating sea mayor de 3"
filtro8 = df[
    (df["review_scores_cleanliness"] < 3) & 
    (df["review_scores_rating"] > 3)
]
filtro8
```

```python
filtro1.to_csv('filtroa')
filtro2.to_csv('filtrob')
filtro3.to_csv('filtroc')
filtro4.to_csv('filtrod')
filtro5.to_csv('filtroe')
filtro6.to_csv('filtrof')
filtro7.to_csv('filtrog')
filtro8.to_csv('filtroh')
```
